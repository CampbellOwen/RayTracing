```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook

num_points = 1000
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlim([-1.0, 1.0])
ax.set_ylim([-1.0, 1.0])
ax.set_zlim([0.0, 1.0])

r1 = np.random.rand(num_points)
r2 = np.random.rand(num_points)
z = np.sqrt(1 - r2)
phi = 2*np.pi*r1

x = np.cos(phi) * np.sqrt(r2)
y = np.sin(phi) * np.sqrt(r2)

ax.scatter(x,y,z)

plt.show()
```

```python
def normalize(x):
    return x / np.linalg.norm(x)

def rand_in_unit_sphere():
    while True:
        point = [(2 * np.random.random_sample()) - 1 for i in range(3)]
        if np.linalg.norm(point) < 1.0:
            return point
        
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.set_xlim([-1.0, 1.0])
ax.set_ylim([-1.0, 1.0])
ax.set_zlim([0.0, 1.0])

points = np.array(list(map(lambda x: normalize(np.array([0,0,1]) + normalize(rand_in_unit_sphere())), np.ones(num_points))))
ax.scatter(points[:, 0], points[:, 1], points[:, 2])

plt.show()

```
